```python
#This part, we go through each player, we check to see if they have played 3 or more seasons. If they have, they are valid and their stats are
#recorded in a dict with their names and empty values


#We then look through the plyer and add values depending on their stats season by season.

#Gets the basketball reference link to all NBA players in its database that have a last name that start with the same letter
#Couldn't find any other way to locate the players given just their names

#allows me to make names with accents, such as 'Luka Doni', read as 'Luka Doncic'



def basketball_reference_link(name):
    last_name = name.split()[-1]
    link = "https://www.basketball-reference.com/players/" + last_name[0] + "/"
    return link

#This checks through all players available until it finds the player we are looking for. Returns the link string for their exact page.
def getPlayerLink(name):
    link = basketball_reference_link(name)
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    tbody = soup.find('tbody')
    for tr in tbody.find_all('tr'):
        player_name = tr.find('th', {'scope': 'row'})
        a_tag = player_name.find('a')
        nametest = a_tag.text
        if unidecode(nametest.lower()) == unidecode(name):
          player_link = a_tag['href']
          return 'https://www.basketball-reference.com' + player_link
    return None

#This serves to get player link. returns None if player does not exist

def getStatsBefore2023(name):
  name = name.lower()
  player_Dict = {
      name : []
  }

  url = getPlayerLink(name)
  td_row=[]
  #if player has never existed, they are invalid. Invalid value of 1.
  if url is None:
    return 1
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  table = soup.find('table', {'id':"per_game"})
  trs = table.find_all('tr', {'class': 'full_table'})
  id_to_remove = "per_game.2023"
  #if player has played less than 3 years (not including 2023), they are invalid. Invalid value of 2.
  if len(trs)<4:
    return 2
  for tr in trs:
    if tr.get('id') == id_to_remove:
        trs.remove(tr)
  for tr in trs:
    player_Dict[name].append([tr.find('td', {'data-stat': 'pts_per_g'}).get_text(),tr.find('td', {'data-stat': 'trb_per_g'}).get_text(),tr.find('td', {'data-stat': 'ast_per_g'}).get_text(),tr.find('td', {'data-stat': 'stl_per_g'}).get_text(),tr.find('td', {'data-stat': 'blk_per_g'}).get_text(),tr.find('td', {'data-stat': 'tov_per_g'}).get_text()])
  return(player_Dict)
  #format of: {'name':[[pts,reb,ast,stl,blks,tov][pts,reb,ast,stl,blks,tov],...]}
```

```python



def basketball_reference_link(name):
    last_name = name.split()[-1]
    link = "https://www.basketball-reference.com/players/" + last_name[0] + "/"
    return link

#This checks through all players available until it finds the player we are looking for. Returns the link string for their exact page.
def getPlayerLink(name):
    link = basketball_reference_link(name)
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    tbody = soup.find('tbody')
    for tr in tbody.find_all('tr'):
        player_name = tr.find('th', {'scope': 'row'})
        a_tag = player_name.find('a')
        nametest = a_tag.text
        if unidecode(nametest.lower()) == unidecode(name):
          player_link = a_tag['href']
          return 'https://www.basketball-reference.com' + player_link
    return None

#This serves to get player link. returns None if player does not exist
```

```python
def player_weighted_ci(player_preds,week,team):
    num_games = get_Player_week(team, week)
    week_stats = []
    for i in range(0,6):
      week_stats.append([player_preds[i][0],player_preds[i][2]-player_preds[i][1]])

    score_multiplier = [1,1.2,1.5,3,3,-1]

    for i in range(0,6):
      week_stats[i][0] = week_stats[i][0] * score_multiplier[i]

    for i in range(0,6):
      week_stats[i][0] = week_stats[i][0] * num_games


    player_stats=week_stats
    # Calculate total expected fantasy points
    total_points = sum(stat[0] for stat in player_stats)

    # Calculate weights for each stat's fantasy points
    weights = [stat[0] / total_points for stat in player_stats]
    # Calculate weighted CI
    weighted_ci = sum(weight * stat[1] for weight, stat in zip(weights, player_stats))
    return [total_points, weighted_ci]

def combinedLeft(player_preds,week,team):
  combined = []
  left = predictedStatline(player_preds)
  right = player_weighted_ci(player_preds,week,team)
  for l in left:
    combined.append(round(l,1))
  for r in right:
    combined.append(round(r,1))
  return combined





```

```python
from scipy.stats import zscore
import numpy as np

def calculate_team_scores(teams):
    # Calculate total points and CI size for each team


    # Extract total points and CI sizes
    total_points = [stats[0] for stats in teams]
    ci_sizes = [stats[1] for stats in teams]

    # Calculate z-scores for total points and CI sizes
    total_points_zscores = zscore(total_points)
    ci_sizes_zscores = zscore(ci_sizes)
    # Calculate scores as the difference between total points z-score and CI size z-score
    scores = total_points_zscores - ci_sizes_zscores
    scores_list = [score for score in scores]

    return scores_list

teams = [[34, 12], [30, 11], [20, 9], [24, 3]]


# Add random dummy stats for the remaining teams


scores = calculate_team_scores(teams)
print(scores)
for i, score in enumerate(scores):
    print(f"Score for Team {i + 1}: {score}")
```
